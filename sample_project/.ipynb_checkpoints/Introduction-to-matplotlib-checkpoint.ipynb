# Matplotlib example workflow

```python
# 0. Import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Preparate the data
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # width and height

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Plot Title",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

# Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x**2);
```

```python
# Use same data to make a Scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nutButterPrices = {
    "Almond": 10,
    "Peanut": 8,
    "Cashew": 12
}
fig, ax = plt.subplots()
ax.bar(nutButterPrices.keys(), nutButterPrices.values());
ax.set(title="Butter Prices",
       ylabel="Price")
```

```python
fig, ax = plt.subplots()
ax.barh(list(nutButterPrices.keys()), list(nutButterPrices.values()));
```

```python
# Make some data for histograms
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
# Plot to each diferent axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nutButterPrices.keys(), nutButterPrices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                   ncols=2,
                   figsize=(10, 5))
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nutButterPrices.keys(), nutButterPrices.values());
ax[1, 1].hist(np.random.randn(1000));
```

# Wich one should I use? (pyplot vs matplotlib OO method)

* When plotting something quickly, okay to use pyplot method
* When plotting something more advanced, use the OO method

```python
heartDisease.head()
```

```python
over50 = heartDisease[heartDisease['age'] > 50]
over50.head()
```

```python
# Pyplot method
over50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
# ax.set_xlim([45, 100])
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over50['age'],
                     y=over50['chol'],
                     c=over50['target']);
ax.set(title='Heart Disease and Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol')

# Add legend
ax.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax.axhline(over50['chol'].mean(),
           linestyle='--');
```

```python
over50.head()
```

```python
# Subplots of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over50['age'],
                      y=over50['chol'],
                      c=over50['target'])

# Customize ax0
ax0.set(title='Heart Disease and Cholesterol Levels',
        ylabel='Cholesterol')

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax0.axhline(y=over50['chol'].mean(),
            linestyle='--')

# Add data to ax1
scatter = ax1.scatter(x=over50['age'],
                      y=over50['thalach'],
                      c=over50['target'])

# Customize ax1
ax1.set(title='Heart Disease and Max Heart Rate',
        xlabel='Age',
        ylabel='Max Heart Rate')

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax1.axhline(y=over50['thalach'].mean(),
            linestyle='--')

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

# Subplots of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over50['age'],
                      y=over50['chol'],
                      c=over50['target'],
                      cmap='winter') # this changes the colour scheme

# Customize ax0
ax0.set(title='Heart Disease and Cholesterol Levels',
        ylabel='Cholesterol')

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax0.axhline(y=over50['chol'].mean(),
            linestyle='--')

# Add data to ax1
scatter = ax1.scatter(x=over50['age'],
                      y=over50['thalach'],
                      c=over50['target'],
                      cmap='viridis')

# Customize ax1
ax1.set(title='Heart Disease and Max Heart Rate',
        xlabel='Age',
        ylabel='Max Heart Rate')

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax1.axhline(y=over50['thalach'].mean(),
            linestyle='--')

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

# Subplots of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over50['age'],
                      y=over50['chol'],
                      c=over50['target'],
                      cmap='viridis') # this changes the colour scheme

# Customize ax0
ax0.set(title='Heart Disease and Cholesterol Levels',
        ylabel='Cholesterol')

# Change x axis limits
ax0.set_xlim([50, 80])

# Change y axis limits
ax0.set_ylim([100, 600])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax0.axhline(y=over50['chol'].mean(),
            linestyle='--')

# Add data to ax1
scatter = ax1.scatter(x=over50['age'],
                      y=over50['thalach'],
                      c=over50['target'],
                      cmap='viridis')

# Customize ax1
ax1.set(title='Heart Disease and Max Heart Rate',
        xlabel='Age',
        ylabel='Max Heart Rate')

# Change x axis limits
ax1.set_xlim([50, 80])

# Change y axis limits
ax1.set_ylim([60, 200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax1.axhline(y=over50['thalach'].mean(),
            linestyle='--')

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```
