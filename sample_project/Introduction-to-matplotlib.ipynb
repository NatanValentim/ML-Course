```python
x = [1, 2, 3, 4]
y = [7, 8, 9, 10]

plt.plot(x, y);
```

```python
# 1st method
fig = plt.figure(); # create a figure
ax = fig.add_subplot(); # add the axes
plt.show()
```

```python
# 2nd method
fig = plt.figure(); # create a figure
ax = fig.add_axes([1, 1, 1, 1]); # add the axes
ax.plot(x, y)
plt.show()
```

```python
# 3rd method (recomended)
fig, ax = plt.subplots()
ax.plot(x, [100, 150, 250, 300]);
type(fig), type(ax)
```

# Matplotlib example workflow

```python
# 0. Import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Preparate the data
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # width and height

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Plot Title",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```
