# Matplotlib example workflow

```python
# 0. Import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Preparate the data
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # width and height

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Plot Title",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

# Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x**2);
```

```python
# Use same data to make a Scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nutButterPrices = {
    "Almond": 10,
    "Peanut": 8,
    "Cashew": 12
}
fig, ax = plt.subplots()
ax.bar(nutButterPrices.keys(), nutButterPrices.values());
ax.set(title="Butter Prices",
       ylabel="Price")
```

```python
fig, ax = plt.subplots()
ax.barh(list(nutButterPrices.keys()), list(nutButterPrices.values()));
```

```python
# Make some data for histograms
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
# Plot to each diferent axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nutButterPrices.keys(), nutButterPrices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                   ncols=2,
                   figsize=(10, 5))
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nutButterPrices.keys(), nutButterPrices.values());
ax[1, 1].hist(np.random.randn(1000));
```

# Plotting from Pandas Dataframe

```python
# Make a Dataframe
carSales = pd.read_csv("car-sales.csv")
carSales
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2020', periods=1000))
ts = ts.cumsum();
ts.plot();
```

```python
fig, ax = plt.subplots()
ax.bar(carSales.Make, carSales['Odometer (KM)']);
```

```python
carSales.Price = carSales.Price.str.replace('[\$\,\.]', '').str[:-2].astype(int)
carSales
```

```python
carSales['Sale Date'] = pd.date_range('1/1/2020', periods=len(carSales))
carSales
```

```python
carSales['Total Sales'] = carSales.Price.cumsum()
carSales
```

```python
fig, ax = plt.subplots()
ax.bar(carSales.Make, carSales['Total Sales']);
```

```python
# Let's plot the total sales
carSales.plot(x="Sale Date", y='Total Sales');
```

```python
carSales.plot(x='Odometer (KM)', y='Price', kind='scatter');
```
